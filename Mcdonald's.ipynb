### Number of items per Category

```python
ax = df.groupby('Category')['Item'].nunique().plot(kind = 'bar')
ax.bar_label(ax.containers[0])
plt.style.use('dark_background')
plt.title("No of items per Catgory",color = 'white',weight = 'bold')
plt.ylabel("Number of Items",color = 'pink',fontsize = 12)
plt.xticks(rotation = 20)
plt.show()
```

There are maximum items in the Coffee and Tea Category and least items in the salads category.

### Avg Calories per Category in the Mcdonald's Menu

```python
ax = df.groupby('Category')['Calories'].mean().plot(kind = 'bar')
ax.bar_label(ax.containers[0])
plt.style.use('dark_background')
plt.title("Avg Calories per Category",color = 'white',weight = 'bold')
plt.ylabel("Avg Calories",color = 'pink',fontsize = 12)
plt.xticks(rotation = 20)
plt.show()
```

From the above graph we can observe that Basically Non-Veg. items has more Calories, specially in Chicken & Fish is the most.If a person want to take some healtheir options in Mcdonald's then it is going to be Beverages, as it has lowest Avg. Calories.

### Avg. Sugar Per Category

```python
ax = df.groupby('Category')['Sugars'].mean().plot(kind = 'barh')
ax.bar_label(ax.containers[0])
plt.style.use('dark_background')
plt.title("Avg Sugar per Category",color = 'white',weight = 'bold')
plt.xlabel("Avg Sugar",color = 'pink',fontsize = 12)
plt.show()
```

Basically the Drinks in the Mcdonald's Menu has the more avg. sugar then others.Smoothies & Shakes tops the least.Snacks & Salads has least suagr content compare to other.

### Avg Protein per Category

```python
ax = df.groupby('Category')['Protein'].mean().plot(kind = 'bar')
ax.bar_label(ax.containers[0])
plt.style.use('dark_background')
plt.title("Avg Protein per Category",color = 'white',weight = 'bold')
plt.ylabel("Avg Protein",color = 'pink',fontsize = 12)
plt.xticks(rotation = 20)
plt.show()
```

Chicken & fish and Beef and pork has most protien content in the entire menu.As expected beverages has the least amount of protien.There is also good amount of protien in salads also.

### Amount of Calories in Crispy Chicken Category in the Menu

```python
# Exploring the amount of calories on Crispy Chicken category
crispy = df[df['Item'].str.contains('Crispy Chicken')]
crispy_cal = pd.DataFrame({'Item': crispy['Item'], 'Calories': crispy['Calories']})
crispy_sort = crispy_cal.sort_values('Calories', ascending=False)
ax = crispy_sort.plot.barh(x='Item', y='Calories', figsize= (11,8))
ax.bar_label(ax.containers[0])
ax.set_title("Total Calories in Crispy Chicken Item Category (in kcal)")
plt.show()
```

### Here we are comparing Avg. Total fat and Avg. Trans Fat

```python
plt.figure(figsize = (16,8))
plt.subplot(1,2,1)
sns.set_palette('viridis')
ax = df.groupby('Category')['Total Fat'].mean().sort_values(ascending = False).plot(kind = 'bar',grid = True)
ax.bar_label(ax.containers[0])
plt.title('Avg. Total Fat per Category')
plt.subplot(1,2,2)
ax = df.groupby('Category')['Trans Fat'].mean().sort_values(ascending = False).plot(kind = 'bar',grid = True)
ax.bar_label(ax.containers[0])
plt.title('Avg. Trans Fat per Category')
plt.suptitle("Avg. Total fat vs Avg. Trans Fat",weight = 'bold',fontsize = 18)
plt.tight_layout()
```

We all know that trans fat is very harmful and we can see that from above that in beef & pork category it is very high and then we have smoothies & shakes after that. Whereas beef & pork have less total fat comparing to others. So, if we compare to others Beef & pork items are unhealthy among all category.

### Comparing Total Calories and Calories coming from fat.

```python
ax = df.groupby('Category')[['Calories','Calories from Fat']].mean().plot(kind = 'bar',width=0.8,figsize = (12,6))
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10)
```

From the graph we can say that Smoothies & Shakes have very high calories and calories from fat is low in smoothies category as compared to others.Calories from fat is very high in Chicken and Breakfast and they have overall high calories.Desserts also has low % of calories from fat in overall calories, similiarly for Coffee & Tea.

### Correlation among all Categories

```python
# Select only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()

# Print the correlation matrix
print(correlation_matrix)

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
